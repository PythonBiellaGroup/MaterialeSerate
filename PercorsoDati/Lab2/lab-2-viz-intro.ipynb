## Matplotlib

For this example we create a dummy dataset to see how to do the visualizations

```python
# Create some data
import numpy as np
rng = np.random.RandomState(0)
x = np.linspace(0, 10, 500)
y = np.cumsum(rng.randn(500, 6), 0)
```

```python
#Let's print an example with matplotlib
figure = plt.figure(figsize=(16,5), dpi=100)
figure = plt.plot(x, y) #color='tab:blue'

plt.gca().set(title="Matplotlib example", xlabel="x", ylabel="y")
plt.legend('Test with Matplotlib', ncol=2, loc='upper left');

plt.show()
```

```python
np.random.seed(42)

# Example data
people = ['Andrea', 'Mario', 'Maria Teresa', 'Cez', 'Pippo']
# y_pos = np.arange(len(people))
y_pos = list(range(0,len(people)))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))
```

```python

# plt.rcdefaults()
fig, ax = plt.subplots(figsize=(16,5), dpi=100)

ax.barh(y_pos, performance, xerr=error, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.set_xlabel('Performance')
ax.set_title('Example with subplot')

# ax.invert_yaxis()  # labels read top-to-bottom

plt.show()
```

## Seaborn

Seaborn is a statistical data visualization library that is actually a wrapper with respect to Matplotlib.  
The advantage of using Seaborn is that it provides more attractive and informative graphics.

```python
import seaborn as sns
sns.set() #Set seaborn as default graphic library
```

```python
figure = plt.figure(figsize=(16,5), dpi=100)
figure = plt.plot(x, y) #color='tab:blue'
plt.gca().set(title="Matplotlib example", xlabel="x", ylabel="y")
plt.legend('Test with Matplotlib', ncol=2, loc='upper left');
plt.show()
```

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])

plt.figure(figsize=(16,5), dpi=100)
for col in 'xy':
    plt.hist(data[col], alpha=0.5)
```

```python
plt.figure(figsize=(16,5), dpi=100)
for col in 'xy':
    sns.kdeplot(data[col], shade=True)
```

Let's create some custom much complex example with a dummy dataset

```python
#First table b1
business_sectors = pd.Series(["Food","Cars", "Prints", "Toys"])
values = pd.Series([0.8,0.7,1,0.4])
b1 = pd.DataFrame({"business sectors": business_sectors, "values": values})
b1
```

```python
#Second table b2
business_sectors = pd.Series(["Metal","Cars"])
values = pd.Series([0.9,0.42])
b2 = pd.DataFrame({"business sectors": business_sectors, "values": values})
b2
```

```python
#Third table b3
business_sectors = pd.Series(["Cars","Electronics","Food"])
values = pd.Series([1,0.92,0.67])
b3 = pd.DataFrame({"business sectors": business_sectors, "values": values})
b3
```

```python
#Let's stick this data together!!!
df = pd.concat([b1,b2,b3],keys=["b1", "b2", "b3"])
df = df.reset_index()
df = df.drop("level_1",axis=1)
df = df.rename({"level_0":"Type"},axis=1)
df
```

Now let's draw an interactive plot with plotly with this data!

```python
#simple plot with a single df
fig = px.bar(b1, x="business sectors", y="values", barmode="group")
fig.show()
```

```python
#Complex plot
#Grafico raggruppato
fig = px.bar(df, x="business sectors", color="Type",
            y='values',
            title="Barplot raggruppato",
            barmode='group',
            height=700,
            )
fig.add_hline(y=0.6, line_width=3, line_color="green", line_dash="dash", 
            annotation_text="Green zone", 
            annotation_position="bottom right")
fig.show()

#Documentation on lines: https://plotly.com/python/horizontal-vertical-shapes/
```

```python
#Same plot with a different group

fig = px.bar(df, x="business sectors", color="Type",
                y='values',
                title="Barplot grouped",
                barmode='group',
                height=700,
                facet_col="Type"
            )

fig.show()
```
