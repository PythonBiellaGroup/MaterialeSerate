**6 Come unire in un unico df (in append) tutti i files .csv di una cartella**

**13 Come controllo se un df ha valori mancanti?**

```python
# input
df = pd.read_csv("AB_NYC_2019.csv")
df
```

```python
# Solution 1
print("Our df has a total of {} null values".format(df.isnull().sum().sum()))
print()

# Solution 2
df.isnull().values.any()
print()

# Solution 3
# A more detailed one
def report_nulls(df):
    '''
    Show a fast report of the DF.
    '''
    rows = df.shape[0]
    columns = df.shape[1]
    null_cols = 0
    list_of_nulls_cols = []
    for col in df.columns:
        null_values_rows = df[col].isnull().sum()
        #Percentuale
        null_rows_pcn = round(((null_values_rows)/rows)*100, 2)
        col_type = df[col].dtype
        if null_values_rows > 0:
            print("The column {} has {} null values. It is {}% of total rows.".format(col, null_values_rows, null_rows_pcn))
            print("The column {} is of type {}.\n".format(col, col_type))
            null_cols += 1
            list_of_nulls_cols.append(col)
    null_cols_pcn = round((null_cols/columns)*100, 2)
    print("The DataFrame has {} columns with null values. It is {}% of total columns.".format(null_cols, null_cols_pcn))
    return list_of_nulls_cols

report_nulls(df)
```

**15 Come sostituire i valori mancanti di colonne numeriche con la loro media?**

```python
# input
df = pd.read_csv("AB_NYC_2019.csv")
# Per estrarre tutte le colonne numeriche!
df.select_dtypes(include=['number'])
#Lavoriamo solo su...
df[["reviews_per_month"]]
```

```python
# Solution
# Somma di tutti i valori nulli
revs_null = df.isnull().sum().sum()
print(revs_null)
# notice that we have filtering the columns  as a list
df[["reviews_per_month"]] = df[["reviews_per_month"]].apply(lambda x: x.fillna(x.mean()))
end_null = df.isnull().sum().sum()
print(end_null)

print("We have got rid of {} null values, filling them with the mean.".format(revs_null - end_null))
df[["reviews_per_month"]]
```

```python
# Solution Q2 - No hard coding
def switch_columns(df, col1=None, col2=None):
    colnames = df.columns.tolist()
    i1, i2 = colnames.index(col1), colnames.index(col2)
    colnames[i2], colnames[i1] = colnames[i1], colnames[i2]
    return df[colnames]

df1 = switch_columns(df, "name", "host_id")
df1
```

**21 Come scambiare due righe in un dataframe?**

```python
# input
df = pd.read_csv("AB_NYC_2019.csv", usecols=["name", "host_name", "neighbourhood_group"])
df.head(5)
```

```python
# Solution
def swap_rows(df, i1, i2):
    a, b = df.iloc[i1, :].copy(), df.iloc[i2, :].copy()
    df.iloc[i1, :], df.iloc[i2, :] = b, a
    return df

swap_rows(df, 1, 2)
```

**25 Come separare una colonna testuale in due colonne separate?**

Split the string column in df to form a dataframe with 3 columns as shown.

```python
# input
df = pd.DataFrame(["STD, City    State",
"33, Kolkata    West Bengal",
"44, Chennai    Tamil Nadu",
"40, Hyderabad    Telengana",
"80, Bangalore    Karnataka"], columns=['row'])

df

'''
Desired Output

0 STD        City        State
1  33     Kolkata  West Bengal
2  44     Chennai   Tamil Nadu
3  40   Hyderabad    Telengana
4  80   Bangalore    Karnataka
'''
```

```python
# we do " ".join(x.split()) to replace multiple spaces to 1 space
# we do split(None, 2, ) to split a string on the second space
# this way we have West Bengal together
# If sep is not specified or is None
# runs of consecutive whitespace are regarded as a single separator and the result will contain no empty strings at the start or end if the string has leading or trailing whitespace
# 2  MaxSplit
# Altro esempio
#'-a-b-c-'.split('-', 2)
#['', 'a', 'b-c-']
df["re"] = df["row"].apply(lambda x: " ".join(x.split()).split(None, 2, ))
df["re"]

new_header = df["re"][0]
values = df["re"][1:]

# I valori sono in una serie di liste: list comprehension per estrarli
d = {new_header[0]:[int(values.iloc[i][0].replace(",", "")) for i in range(len(values))], \
     new_header[1]:[values.iloc[i][1] for i in range(len(values))], \
     new_header[2]:[values.iloc[i][2] for i in range(len(values))]}
d
# create a pandas DF from a dict
new_df = pd.DataFrame(d)
new_df
```
